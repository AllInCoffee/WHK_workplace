```python
import math
import numpy.ma as ma
import numpy as np
import matplotlib.pyplot as plt
# trajectories of I(n) and dI(n)
def fig2(k_T, k, x, delta):
    '''ODE system modeling a single optical nonlinear resonator
    Args:
        k_T: w0/Q, total linewidth of the resonator given its resonance angular frequency Ï‰0 and loaded
quality factor Q
        k_W: waveguid linewidth
        k_L: loss linewidth
        k_T = k_W + k_L
        -----
        delta: delta = w0-w, the detuning of the resonance frequency from the drive field
        x: Chi, dispersive non-linear coefficient, strength of the optical Kerr effect, defined as the angular
frequency shift per photon inside the resonator.
        u: coherent drive/input amplitude, normalized to sqr(# Photons/time)
           input flux I = |u|^2
        z: complex mode amplitude of Kerr-nonlinear resonator, represents dynamics of a single resonant mode
           number of photons n := |z|^2
    return: n vs u
    '''
    a = x**2 / k
    b = 2*delta*x /k
    c = (k_T**2 /4 + delta**2)/k
    d = 0
    p0=153.63*math.sqrt(abs(x))
    lower_limit = 0
    upper_limit = 400
    num_pts = 200
    n = np.linspace(lower_limit, upper_limit, num_pts)
    I = np.polyval([a,b,c,d], n)
    dI = np.polyval([3*a,2*b,c], n)
    u = [math.sqrt(i) for i in I]
    
    plt.xlabel('Pump Amplitude p')
    plt.ylabel('Energy Cavity n')
    Delta_th= -math.sqrt(3)/2.0 * K_T
    m = dI <= 0
    plt.plot(np.ma.array(u, mask=m), np.ma.array(n, mask=m), '-', lw=1.5, label="$\Delta = {:.1f} \Delta_{{th}}$".format(delta/Delta_th))
    m = dI > 0
    plt.plot(np.ma.array(u, mask=m), np.ma.array(n, mask=m), '--', lw=1.5)
    
    
        
```

```python
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt
def fig4(k, k_l, delta, re_r, im_r, x):
    r=complex(re_r,im_r)
    K = k * abs(x)
    K_t=(k+k_l)*abs(x)
    K_feedback=((1+r)/(1-r)).real*k*abs(x)
    K_t_feedback= (((1+r)/(1-r)).real*k+k_l)*abs(x)
    Delta=delta*x
    Delta_feedback=Delta+ (r/(1-r)).imag *k*abs(x)
    # range of s_tilde_square where bifurcation is
    s_tilde_square_minus= -4*Delta/(3*x) - math.sqrt(4*(Delta**2)/(9*(abs(x)**2))-(K_t**2)/(3*(abs(x)**2)))
    s_tilde_square_plus= -4*Delta/(3*x) + math.sqrt(4*(Delta**2)/(9*(abs(x)**2))-(K_t**2)/(3*(abs(x)**2)))
    # relevant p for s_tilde
    p_square_plus = ((x**2*s_tilde_square_plus**3)/4+Delta_feedback*x*s_tilde_square_plus**2+((K_t_feedback**2)/4+Delta_feedback**2)*s_tilde_square_plus)/K_feedback
    p_square_minus = ((x**2*s_tilde_square_minus**3)/4+Delta_feedback*x*s_tilde_square_minus**2+((K_t_feedback**2)/4+Delta_feedback**2)*s_tilde_square_minus)/K_feedback
    p=[]
    g1=[]
    g2=[]
    g1_new=[]
    g2_new=[]
    arg1=[]
    lower_limit = 0.001
    upper_limit = 600
    
    p0=153.63*math.sqrt(abs(x))
    s_tilde_square=np.arange(lower_limit, upper_limit, 2)  # step size decides resolution! e.g. being changed to 10 didn't look good
    
    for i in s_tilde_square:
        p_square=((x**2*i**3)/4+Delta_feedback*x*i**2+((K_t_feedback**2)/4+Delta_feedback**2)*i)/K_feedback
        p.append(math.sqrt(p_square)/p0)
        # formel 43 to get complex s_tilde, assuming p real
        temp1= abs(1-r)*1j/(1-r) * math.sqrt(p_square) * math.sqrt(K_feedback)
        temp2=-(0.5*K_t_feedback + 1j*Delta_feedback + 1j*x*i*0.5)
        s_tilde= temp1/temp2
        # linear gain of p
        denominator=K_t**2/4.0 + (Delta+x*i)**2.0 - abs(x)**2*i**2/4.0
        g1.append(K*(+K_t/2.0 + 1j*Delta + 1j*x*i)/denominator +1)
        #g2.append( (1j*K*x*i*0.5)/denominator)  #wrong, s_tilde**2 is complex 
        g2.append( (1j*K*x*s_tilde**2*0.5)/denominator)
            
        
        if i <s_tilde_square_minus or i > s_tilde_square_plus:  # outside splitting regime, g1 = -g-, g2= g+
            # linear gain of u_tilde # from formel above (39)
            g1_new.append(-K*(-K_t/2.0 + 1j*Delta+ 1j*x*i)/denominator-1)
            g2_new.append( 1j*K*x*s_tilde**2*0.5/denominator)
        else:                                                   # inside splitting regime, g1 = g-, g2= - g+
            g1_new.append(K*(-K_t/2.0 + 1j*Delta+ 1j*x*i)/denominator+1)
            g2_new.append( -1j*K*x*s_tilde**2*0.5/denominator)
            
            
    #plus=[abs(abs(i)+abs(j)) for i, j in zip(g1,g2)]
    #minus=[abs(abs(i)-abs(j)) for i, j in zip(g1,g2)] 

    plus=[abs(abs(i)+abs(j)) for i, j in zip(g1_new,g2_new)]
    minus=[abs(abs(i)-abs(j)) for i, j in zip(g1_new,g2_new)]

    arg1=[np.angle(i) for i in g1_new] 
    arg2=[cmath.phase(i) for i in g2_new]       
    
    # plot 
    plt.close('all')
    
    plt.subplot(2,1,1)
    plt.title('Linear Transferfunction')
    plt.plot(p,[20*math.log10(k) for k in plus],'r', label=r'$||T_{vu}|+|T_{vu^*}||$')
    plt.plot(p,[20*math.log10(k) for k in minus],label=r'$||T_{vu}|-|T_{vu^*}||$')
    plt.axvline(x=math.sqrt(p_square_minus)/p0, color='black', linestyle='dashed',label=r'$p_{th}$')
    plt.ylim(-60,60)
    plt.xticks(np.arange(0, 2.3, step=0.5))
    plt.legend()
    #plt.axvline(x=math.sqrt(p_square_plus)/p0, color='black', linestyle='dashed')
    plt.ylabel('Quadrature Gain [dB]')
    plt.grid(True)
    
    plt.subplot(2,1,2)
    plt.plot(p,arg1,'r', label=r'$\Phi = argT_{vu}$')
    plt.plot(p,arg2, label=r'$\Psi = argT_{vu^*}$')
    plt.axvline(x=math.sqrt(p_square_minus)/p0, color='black', linestyle='dashed',label=r'$p_{th}$')
    plt.ylabel('Phase [rad]')
    plt.grid(True)
    plt.xlabel(r'Pump Amplitude p [$p_0$]')
    plt.yticks(np.arange(-4, 4, step=1))
    plt.xticks(np.arange(0, 2.3, step=0.5))
    plt.tight_layout()
    plt.legend()
    plt.show()

    
```
